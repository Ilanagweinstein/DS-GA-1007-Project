```python
from bs4 import BeautifulSoup
import requests

#Top 250 movies on IMDb
url = "https://www.imdb.com/search/title/?groups=top_250&sort=user_rating"
```

```python
resp = requests.get(url)
resp
```

```python
#All info from webpage scraped
soup = BeautifulSoup(resp.text,'html.parser')
soup
```

```python
#Obtaining only h3 headings which is where the movie titles are stored
scraped_titles = soup.find_all('h3',{"class":"lister-item-header"})
scraped_titles
```

```python
#Stripping for just the movie title
movies = []
for x in scraped_titles:
    for y in x.find_all('a'):
        movies.append(y.text)
```

```python
movies
```

```python
#Stripping for just the movie year
years = []
for x in scraped_titles:
    for y in x.find_all('span'):
        years.append(y.text)
```

```python
years = years[1::2]
```

```python
len(years)
```

```python
year = []

for i in len(range(years)):
    i.strip("(")
    i.strip(")")
    year.append(i)
    
```

```python
a = []
b = []

for i in range(len(years)):
    a.append(years[i].strip("("))
    b.append(a[i].strip(")"))
```

```python
years = [];
for _ in b:
    years.append(int(_));

    
years
```
